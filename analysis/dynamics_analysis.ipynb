## Dynamics Analysis

Here is the analysis of the physical dynamics of the simulation. This includes pressure field, solid velocity, and mass flux fields.

First we import the necessary modules, and initialize the analyzers.
A simulation data file is passed to each analyzer, which contains the fields to be analyzed.

```python
import sys

sys.path.append('/Users/rileymcnamara/CODE/2025/SCIE3212/')

from analysis.pressure_analysis import PressureAnalyzer
from analysis.solid_velocity_analysis import SolidVelocityAnalyzer
from analysis.tumor_analysis import TumorAnalyzer
from src.models.SCIE3121_model import SCIE3121_MODEL
from src.models.initial_conditions import SphericalTumor
from src.utils.utils import SCIE3121_params

simulation_data = '/Users/rileymcnamara/CODE/2025/SCIE3212/data/project_model_test_sim_data.npz'

model = SCIE3121_MODEL(
        grid_shape=(100, 100, 100),
        dx=0.2,
        dt=0.001,
        params=SCIE3121_params,
        initial_conditions=SphericalTumor(grid_shape=(150, 150, 150), radius=5, nutrient_value=1.0),)
    

pressure_analyzer = PressureAnalyzer(simulation_data)
solid_velocity_analyzer = SolidVelocityAnalyzer(simulation_data)
tumor_analyzer = TumorAnalyzer(simulation_data)
```
